In this notebook I visualize hand-labeled points from mesh map (also included in repo)

In file facemesh_2d_map_points.npy all 2d coordinates of every point provided

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import cv2
face_image = cv2.imread("../mesh_map.jpg")[..., ::-1]
```

```python
all_points = np.load("facemesh_2d_map_points.npy")
```

```python
face_image.shape, all_points.max()
```

```python
plt.figure(figsize=(20, 12))

im = face_image.copy()
for pt in all_points:
    new_pt = np.array(pt)
    new_pt = new_pt.astype(int)
    im = cv2.circle(im, tuple(new_pt), 10, (100, 100, 200), 10)
plt.imshow(im)
```
